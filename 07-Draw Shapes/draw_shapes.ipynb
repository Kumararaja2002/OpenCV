```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Create a empty canvas
height, width = 500, 700
blue = (255, 127, 0)
canvas = np.full((height, width, 3), blue, dtype=np.uint8)

plt.imshow(cv2.cvtColor(canvas, cv2.COLOR_BGR2RGB))
```

### Draw to a line

```python
# cv2.line(canvas, starting-point-coordinate (50, 50), ending-point-coordinate (650, 650), color-of-the-shape(255, 255, 255), weight-of-the-shape(5))
cv2.line(canvas, (50, 50), (650, 450), (255, 255, 255), 5)

plt.imshow(cv2.cvtColor(canvas, cv2.COLOR_BGR2RGB))
```

### Draw a rectangle

```python
cv2.rectangle(canvas, (100, 100), (300, 300), (0, 255, 255), -1)

plt.imshow(cv2.cvtColor(canvas, cv2.COLOR_BGR2RGB))
```

### Draw a circle

```python
cv2.circle(canvas, (500, 150), 100, (0, 0, 255), 5)

plt.imshow(cv2.cvtColor(canvas, cv2.COLOR_BGR2RGB))
```

### Draw an Ellipse

```python
cv2.ellipse(canvas, (350, 350), (100, 50), 45, 0, 360, (0, 255, 0), 3)

plt.imshow(cv2.cvtColor(canvas, cv2.COLOR_BGR2RGB))
```

### Draw a polygon

```python
# Xn, Yn, where n>2

points = np.array([[150, 400], [300, 400], [225, 300]], np.int32)
new_reshaped_point = points.reshape((-1, 1, 2))

print(new_reshaped_point)


cv2.polylines(canvas, [new_reshaped_point], True, (255, 0, 255), 3) # If we provide "False", we won't be getting any closed shape.w

plt.imshow(cv2.cvtColor(canvas, cv2.COLOR_BGR2RGB))
```
