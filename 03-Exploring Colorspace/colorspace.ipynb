```python
import cv2
```

```python
bgr_image = cv2.imread('./mountain.jpg')
```

```python
# Convert to grayscale
grayscale_image = cv2.cvtColor(bgr_image, cv2.COLOR_BGR2GRAY)
cv2.imwrite('./grayscale_image.jpg',grayscale_image)
```

```python
# Covert to HSV
hsv_image = cv2.cvtColor(bgr_image, cv2.COLOR_BGR2HSV)
cv2.imwrite('./hsv_image.jpg',hsv_image)
```

```python
# Convert to LAB
lab_image = cv2.cvtColor(bgr_image, cv2.COLOR_BGR2LAB)
cv2.imwrite('./lab_image.jpg',lab_image)
```

```python
# Convert to RGB
rgb_image = cv2.cvtColor(bgr_image, cv2.COLOR_BGR2RGB)
cv2.imwrite("./rgb_image.jpg", rgb_image)
```

In OpenCV, by default images are stored in BGR because of historical practice. But while using it convert to RGB defaulty.
