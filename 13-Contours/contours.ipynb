```python
import cv2
import matplotlib.pyplot as plt
```

```python
image = cv2.imread('./shapes.jpg')

image_rgb = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)

image_gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
```

```python
plt.imshow(image_rgb)
```

```python
# Edge Detection method
gray = cv2.Canny(image_gray, 80, 150)

plt.imshow(gray, cmap='grey')
```

```python
contours, hierarchy = cv2.findContours(gray, cv2.RETR_LIST, cv2.CHAIN_APPROX_SIMPLE)

# cv2.RETR_LIST --> MODE - hierarchy relationship between contours
# cv2.CHAIN_APPROX_SIMPLE --> METHOD - contours points are stored.

contour_image = image.copy()
cv2.drawContours(contour_image, contours, -1, (0, 255, 0), 5)

plt.figure(figsize=(10,5))
plt.subplot(1,3,1)
plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
plt.title("Original image")

plt.subplot(1,3,2)
plt.imshow(gray, cmap='grey')
plt.title("maks")

plt.subplot(1,3,3)
plt.imshow(cv2.cvtColor(contour_image, cv2.COLOR_BGR2RGB))
plt.title("Contours image")

plt.show()
```

```python
# Threshold method
retval, binary = cv2.threshold(image_gray, 200, 255, cv2.THRESH_BINARY)

plt.imshow(gray, cmap='grey')
```

```python
contours, hierarchy = cv2.findContours(binary, cv2.RETR_LIST, cv2.CHAIN_APPROX_SIMPLE)

# cv2.RETR_LIST --> MODE - hierarchy relationship between contours
# cv2.CHAIN_APPROX_SIMPLE --> METHOD - contours points are stored.

contour_image = image.copy()
cv2.drawContours(contour_image, contours, -1, (0, 255, 0), 5)

plt.figure(figsize=(10,5))
plt.subplot(1,3,1)
plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
plt.title("Original image")

plt.subplot(1,3,2)
plt.imshow(binary, cmap='grey')
plt.title("maks")

plt.subplot(1,3,3)
plt.imshow(cv2.cvtColor(contour_image, cv2.COLOR_BGR2RGB))
plt.title("Contours image")

plt.show()
```

```python
contour_image = image.copy()

for contour in contours:
    x, y, w, h = cv2.boundingRect(contour)
    cv2.rectangle(contour_image, (x,y), (x+w, y+h), (0,0,255), 2)


# cv2.drawContours(contour_image, contours, -1, (0, 255, 0), 5)

plt.figure(figsize=(10,5))
plt.subplot(1,3,1)
plt.imshow(image_grey, cmap='grey')
plt.title("Original image")

plt.subplot(1,3,2)
plt.imshow(mask, cmap='grey')
plt.title("maks")

plt.subplot(1,3,3)
plt.imshow(cv2.cvtColor(contour_image, cv2.COLOR_BGR2RGB))
plt.title("Contours image")

plt.show()
```
